# 1. Import Library

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

# 2. Import Data

```python
df= pd.read_csv('daily_min_temp.csv',index_col="Date",parse_dates=["Date"])
df
```

```python
df=df.resample('M').mean()
df
```

# 3. Data Prepocessing

```python
cols=list()
```

```python
for i in range(1,0,-1):
    cols.append(df.shift(i))
cols
```

```python
for i in range(2,0,-2):
    cols.append(df.shift(i))
cols
```

```python
for i in range(3,0,-3):
    cols.append(df.shift(i))
cols
```

```python
for i in range(0,1):
    print(i)
    cols.append(df.shift(i))
cols
```

```python
agg=pd.concat(cols,axis=1)
agg
```

```python
agg.dropna(inplace=True)
agg
```

# 4. Train Test Split

```python
train,test=agg.iloc[:-10,:],agg.iloc[-10:,:]
```

```python
train
```

```python
test
```

```python
trainX=train.iloc[:,:-1]
trainX
```

```python
trainy=train.iloc[:,-1]
trainy
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.linear_model import SGDRegressor
from sklearn.svm import SVR

model=KNeighborsRegressor(n_neighbors=8)
model.fit(trainX,trainy)
```

# 6. Prediction

```python
actual=list()
predictions=list()
for i in range(len(test)):
    testX,testy=test.iloc[i,:-1],test.iloc[i,-1]
    pred=model.predict([testX])
    print("hasil prediksi :"+str(pred[0]))
    predictions.append(pred[0])
    actual.append(testy)
    print("actual: "+str(testy)+" predicted= "+str(pred[0]))
```

```python
from sklearn.metrics import mean_absolute_error
error=mean_absolute_error(actual,predictions)
print(error)
```

```python
plt.plot(actual, label="actual")
plt.plot(predictions, label="predictions")
plt.legend()
plt.show()
```

# 7. DataFrame New

```python
df_visual = pd.DataFrame({
    'date':test.reset_index()["Date"].to_list(),
    'actual':actual,
    'predictions':predictions
})
df_visual
```

```python
df_visual.set_index('date',inplace=True)
df_visual
plt.figure(figsize=(15,5))
plt.plot(df_visual.actual,"b-",label="actual")
plt.plot(df_visual.predictions,"r-",label="predictions")
plt.legend()
plt.show()
```

# 8. Predicted

```python
result=pd.concat([df,df_visual],axis=1)
result=result.drop(columns='actual')
result
```

```python
plt.figure(figsize=(15,5))
plt.plot(result.Temp,"b-",label="actual")
plt.plot(result.predictions,"r-",label="predictions")
plt.legend()
plt.show()
```
